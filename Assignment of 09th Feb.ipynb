```python
Answer 1:
    
    class Vehicle:
        def __init__(self, name_of_vehicle, max_speed, average_speed):
            self.name_of_vehicle = name_of_vehicle
            self.max_speed = max_speed
            self.average_speed = average_speed

    # Creating instances of the Vehicle class
    car = Vehicle("Car", 200, 80)
    bike = Vehicle("Bike", 150, 60)

    # Accessing instance variables
    print("Car details:")
    print("Name:", car.name_of_vehicle)
    print("Max Speed:", car.max_speed)
    print("Average Speed:", car.average_speed)

    print("\nBike details:")
    print("Name:", bike.name_of_vehicle)
    print("Max Speed:", bike.max_speed)
    print("Average Speed:", bike.average_speed)
    
    
 

Answer 2:
    
    class Vehicle:
        def __init__(self, name_of_vehicle, max_speed, average_speed):
            self.name_of_vehicle = name_of_vehicle
            self.max_speed = max_speed
            self.average_speed = average_speed

    class Car(Vehicle):
        def seating_capacity(self, capacity):
            return f"{self.name_of_vehicle} has a seating capacity of {capacity}."

    # Creating instances of the Car class
    car1 = Car("Sedan", 180, 70)
    car2 = Car("SUV", 220, 75)

    # Using the seating_capacity method
    print(car1.seating_capacity(5))  # Output: Sedan has a seating capacity of 5.
    print(car2.seating_capacity(7))  # Output: SUV has a seating capacity of 7.


    
    
Answer 3:
    
    Multiple inheritance is a feature in object-oriented programming where a class can inherit attributes and methods 
    from more than one parent class. This allows a subclass to inherit behaviors and characteristics from multiple parent
    classes. In multiple inheritance, a class can inherit from two or more base classes, and it can have access to the 
    attributes and methods of all its parent classes.

    Here's an example of multiple inheritance in Python:
    
    class Engine:
        def start(self):
            print("Engine started")

    class Electric:
        def charge(self):
            print("Charging the electric vehicle")

    class HybridCar(Engine, Electric):
        def __init__(self, name):
            self.name = name

    # Creating an instance of HybridCar
    hybrid_car = HybridCar("HybridCar123")

    # Accessing methods from both parent classes
    hybrid_car.start()   # Output: Engine started
    hybrid_car.charge()  # Output: Charging the electric vehicle
    
    
    

Answer 4:
    
    Getter and setter methods in Python are used to control the access and modification of class attributes (instance 
    variables). They provide a way to retrieve (get) and update (set) the values of private attributes while encapsulating
    the internal implementation details. This helps in maintaining data integrity, applying validation, and providing a 
    controlled interface for interacting with class attributes.

    Here's an example of a class with getter and setter methods:
    
    class Person:
        def __init__(self, name, age):
            self._name = name 
            self._age = age    
    
        # Getter method for 'name'
        def get_name(self):
            return self._name
    
        # Setter method for 'name'
        def set_name(self, new_name):
            if isinstance(new_name, str):
                self._name = new_name
            else:
                print("Invalid name format")
    
        # Getter method for 'age'
        def get_age(self):
            return self._age
    
        # Setter method for 'age'
        def set_age(self, new_age):
            if isinstance(new_age, int) and new_age >= 0:
                self._age = new_age
            else:
                print("Invalid age format")

    # Creating an instance of the Person class
    person = Person("Alice", 25)

    # Using getter and setter methods
    print("Initial name:", person.get_name())  # Output: Initial name: Alice
    person.set_name("Bob")
    print("Updated name:", person.get_name())  # Output: Updated name: Bob

    print("Initial age:", person.get_age())    # Output: Initial age: 25
    person.set_age(30)
    print("Updated age:", person.get_age())    # Output: Updated age: 30
    
    
    
    
Answer 5:
    
    Method overriding in Python occurs when a subclass provides a new implementation for a method that is already defined
    in its parent class. The overridden method in the subclass has the same name, parameters, and return type as the method
    in the parent class. This allows the subclass to customize the behavior of the method while inheriting the general
    structure from the parent class.

    Here's an example to demonstrate method overriding in Python:
    
    class Animal:
        def speak(self):
            print("Animal speaks")

    class Dog(Animal):
        def speak(self):
            print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

# Creating instances
animal = Animal()
dog = Dog()
cat = Cat()

# Calling the speak method on different instances
animal.speak()  # Output: Animal speaks
dog.speak()     # Output: Dog barks
cat.speak()     # Output: Cat meows


```
